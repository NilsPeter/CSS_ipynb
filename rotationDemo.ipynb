```python
import thLib
%pylab inline
```

# Generate dummy head movements

```python
t = arange(0,10,0.1)
gyr = vstack( (zeros_like(t), 0.1*sin(t), zeros_like(t)) ).T
plot(gyr)
legend(['x', 'y', 'z'])
```

# Rotate the sensor 5 deg clockwise

```python
Rmat = thLib.rotmat.R1(-5)
#print(Rmat)

gyrRotated = Rmat.dot(gyr.T).T
plot(gyrRotated)
legend(['x', 'y', 'z'])
```

## Implemented as Matrices

```python
R = matrix(Rmat)
omega = matrix(gyr)
rotated = (R * gyr.T).T
plot(rotated)
```

## Rotation with quaterions

```python
q = [-sin(deg2rad(5)/2), 0, 0]
rotated = thLib.vector.rotate_vector(gyr, q)
plot(rotated)
    
```
